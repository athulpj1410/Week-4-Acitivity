```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Telco-Customer-Churn.xlsx")
df.head()
```

```python
#Compare churn count with respect to gender
```

```python
plt.figure()

#sns.boxplot(x = "Churn", y ="gender",hue = "gender", data = df2)

#df.set_index('gender', inplace=True)
df.groupby(['gender','Churn'])['Churn'].count().plot(kind='bar',legend=True)

plt.show()
```

```python
#Most of females and males said No and only some said yes to Churn
```

```python
#Count of female senior citizens

a=df[(df["SeniorCitizen"]>=1.0)]
plt.figure(figsize=(10,6))
a.groupby('gender')['gender'].count().plot(kind='bar')

plt.title("Senior Citizen")
plt.show()
```

```python
#The count of female senior citizens are less compared to total females 
```

```python
# tenure with total charges
a=pd.read_excel("Telco-Customer-Churn.xlsx",na_values=[" "])
plt.figure(figsize=(10,6))
sns.lineplot(x='tenure',y='TotalCharges',data = a)
plt.title("Tenure v/s Total Charges",fontsize=30)
plt.show()
```

```python
# Tenure is directly proportional to total charges in general
```

```python
#Contract - Senior citizen

a = df[(df["SeniorCitizen"]>=1.0)]
plt.figure(figsize=(10,6))
a.groupby('Contract')['SeniorCitizen'].count().plot(kind='bar',legend=True)
plt.title("Contract v/s Senior Citizen",fontsize = 24)
plt.ylabel("Senior Citizen Count")
plt.show()
```

```python
# Most of the senior Citizen prefers month to month contract
```

```python
# Payment Method

a=pd.read_excel("Telco-Customer-Churn.xlsx",na_values=[" "])
plt.figure(figsize=(10,6))
plt.title("Payment Method",fontsize=24)
a.groupby('PaymentMethod')['PaymentMethod'].count().plot(kind='bar')
plt.ylabel("Count")
plt.show()
```

```python
# Most prefers Electronic Check
```
